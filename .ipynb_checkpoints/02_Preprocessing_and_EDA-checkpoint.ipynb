```python
# We define a function to remove punctuation
def no_punc(input_string):
    return re.sub("[^a-zA-Z]", " ", input_string.lower())
```

```python
def lemmatize(input_string):
    
    # Lemmatise each word of text
    tokenizer = RegexpTokenizer(r'\w+')
    tokens = tokenizer.tokenize(input_string)
    lemmatizer = WordNetLemmatizer()
    tokens_lem = [lemmatizer.lemmatize(i) for i in tokens]

    # Remove stopwords
    stops = set(stopwords.words('english'))
    meaningful_words = [w for w in tokens_lem if not w in stops]
    
    # Join the words back into one string separated by space, and return the result
    return(" ".join(meaningful_words))
```

```python
X_train['text_no_punc'] = [no_punc(i) for i in list(X_train['text'])]
```

```python
%%time
X_train['lemm_text'] = [lemmatize(i) for i in list(X_train['text_no_punc'])] # new column to hold lemmatised text
```

## Exploratory Data Analysis (EDA) on Pre-processed Data

### Word Cloud for Class 7 Variations

```python
def word_count(input_string):
    tokenizer = RegexpTokenizer(r'[A-Za-z0-9]{2,}')
    words = tokenizer.tokenize(input_string)
    counter.update(words)
```

```python
# Create a word frequency dictionary from the stemmed text for class 7 variations
counter = Counter()
for input_string in X_train[y_train.values==7]['lemm_text']:
    word_count(input_string)
class_7_word_freq = dict(counter.most_common())
```

```python
# Determine size of word frequency dictionary
len(counter)
```

```python
# Generate a word cloud image
wordcloud = WordCloud(max_font_size=55).generate_from_frequencies(class_7_word_freq)
plt.figure(figsize=(8,8))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud for Class 7 Variations")
plt.show()
```

<mark>Need to assign more stop words: mutation, figure, tumor, cell, patient</mark>

### Histogram of 20 most frequent words in class 7 variations

```python
labels, values = zip(*class_7_word_freq.items())
# sort your values in descending order
indSort = np.argsort(values)[::-1]
# rearrange your data
labels = np.array(labels)[indSort][:20]
values = np.array(values)[indSort][:20]
indexes = np.arange(len(labels))
```

```python
plt.figure(figsize=(12,5))
sns.barplot(indexes, values);
plt.xticks(indexes, labels);
plt.xlabel("Word in Descriptive Text")
plt.ylabel("Number of occurrences")
plt.title("Histogram of Word Frequencies in Class 7 Variations");
```
